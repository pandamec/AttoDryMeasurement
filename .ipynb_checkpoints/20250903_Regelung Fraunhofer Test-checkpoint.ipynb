## III. Steps Automation

### Functions

```python
def defineStepsScan(Ti,Tf):
    T_range =[]
    T       =Ti
    T_range.append(T)
    
    while T<Tf:

        if T<20:
            dT=2
        
        elif T<50:
            dT=2
            
        elif T<150:
            dT=10
        
        else:
            dT=25

        T=T+dT
        if T<Tf:
            T_range.append(T)
    
    #if T>Tf:
    T_range.append(Tf)
    
    
    return T_range
```

### Procedure

```python
## Temperatures from Attocube
TS = Atto.sample.getTemperature()  # Sample Temperature [K]
TC = Atto.tboard.getTemperature(0) # Coldplate temperature [K]
Q  = Atto.sample.getHeaterPower()  # Sample heater [W]
print(TS)
print(TC)
```

```python
## Parameters for calibration

filename= 'R250815-Kryostat.csv'
Ti = round((TS/2)+0.5)*2 # Initial temperature rounded to the first value multiple of 2 [K]
#Ti = round(TS)+1 # Initial temperature rounded [K]
#Ti=2
Tf = 50           # Final temeprature [K]
t_step=15*60       # Time for each step [s]
T_range = defineStepsScan(Ti,Tf)
#T_range=[60, 85, 100 ,125 ,150]
Time_estimated=(len(T_range)-1)*t_step
print("The estimated time to complete the test is " + str(Time_estimated/3600) + " hours")
print(T_range)
```

```python
Ti
```

```python

## Initial values
log=[]
TS=Atto.sample.getTemperature()
TC=Atto.tboard.getTemperature(0)
Q =Atto.sample.getHeaterPower()
R_sensor= Atto.sample.getResistance()
    
data={
            'Time[s]': datetime.now().timestamp(),
            'Cold-Plate[K]': TC,
            'Sensor Resistance[Ohm]': R_sensor,
            'Sample Temperature[K]': TS,
            'Sample Heat[W]': Q,
            'Set Temperature[K]': TS 
        }
    
log.append(data)
df_log=pd.DataFrame(log)
df_log.to_csv(filename, index=False)
print(data)
```

```python
## Calibration

for i in T_range:

    Atto.sample.setSetPoint(i)
    Atto.sample.startTempControl()
    time.sleep(t_step)

    TS=Atto.sample.getTemperature()
    TC=Atto.tboard.getTemperature(0)
    Q =Atto.sample.getHeaterPower()
    R_sensor= Atto.sample.getResistance()
    
    data={
            'Time[s]': datetime.now().timestamp(),
            'Cold-Plate[K]': TC,
            'Sensor Resistance[Ohm]': R_sensor,
            'Sample Temperature[K]': TS,
            'Sample Heat[W]': Q,
            'Set Temperature[K]': i 
        }
    
    log.append(data)
    df_log=pd.DataFrame(log)
    df_log.to_csv(filename, index=False)
    print(data)

print(log)


Atto.sample.stopTempControl()
```
